```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./matches.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df['player_of_match'].value_counts()[:10]
```

```python
plt.figure(figsize=(7,7))
plt.bar(list(df['player_of_match'].value_counts()[:5].keys()), list(df['player_of_match'].value_counts()[:5]), color=['red', 'green', 'orange', 'yellow', 'blue'])
plt.show()
```

```python
df['result'].value_counts()
```

```python
df['toss_winner'].value_counts()
```

```python
batting_first = df[df['win_by_runs']!=0]
```

```python
batting_first.head()
```

```python
plt.figure(figsize=(7,7))
plt.hist(batting_first['win_by_runs'])
plt.ylabel("No. Of Matches")
plt.xlabel("No. Of Runs")
plt.show()
```

```python
plt.figure(figsize=(15,7))
plt.bar(list(batting_first['winner'].value_counts()[0:5].keys()),list(batting_first['winner'].value_counts()[0:5]), color=['orange', 'cyan', 'green', 'red', 'purple'])
plt.ylabel("No of wins")
plt.xlabel("Teams")
plt.show()
```

```python
plt.figure(figsize=(10, 10))
plt.pie(list(batting_first['winner'].value_counts()), labels=list(batting_first['winner'].value_counts().keys()), autopct='%0.1f%%')
plt.show()

```
